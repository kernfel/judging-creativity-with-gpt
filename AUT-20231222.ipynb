# Exclude incomplete answers

```python
data = pd.read_excel(f'gpt-ratings/Aut_20231222.xlsx', index_col=0)
```

```python
q = 'The string "{Answer}" is a participant\'s answer to the question of alternative uses of the item "{Question}". Input was cut off after a time limit. Does the answer, which may be extremely brief, appear to have been cut off? Yes/No'
```

```python
import numpy as np
firsthundred = np.ones(100, bool)
for i in (1, 13, 37, 40, 50, 55, 62, 64, 91):#, 73):
    firsthundred[i] = False
```

```python
data['answer_complete'] = 0
def process(rq, completion, **kwargs):
    a = completion['choices'][0]['message']['content']
    answer_complete = a.startswith('No')
    if answer_complete:
        data.loc[rq['i'], 'answer_complete'] += 1
        
    if firsthundred[rq['i']] != answer_complete:
        print(rq['i'], data.loc[rq['i'], 'Question'], data.loc[rq['i'], 'Answer'])
        print(a)
```

```python
requests = [{'messages': [{'role': 'user', 'content': q.format(**row)}], 'process': process, 'i': i} for i, row in data[:100].iterrows()]
await rate.entrypoint(requests, model='gpt-4-1106-preview', max_tokens=1, temperature=0)
```

```python
for i in np.flatnonzero(~firsthundred):
    print(i, data.loc[i, 'Answer'])
```

### Incompleteness summary:
- Two statements (22, 47) are false positives (complete answers, marked incomplete);
- Two statements (55, 64) are false negatives, but understandably so;
- One statement (73) is marked incomplete, but may simply be extremely brief
- Excluding answers that end in a full stop leaves 1 possible false positive (73).

```python
data['answer_complete'] = 0
def process(rq, completion, **kwargs):
    a = completion['choices'][0]['message']['content']
    answer_complete = a.startswith('No')
    if answer_complete:
        data.loc[rq['i'], 'answer_complete'] = 1
```

```python
requests = []
for i, row in data.iterrows():
    if row['Answer'].endswith('。'):
        row.loc[i, 'answer_complete'] = 1
    else:
        requests.append({'messages': [{'role': 'user', 'content': q.format(**row)}], 'process': process, 'i': i})
len(data), len(requests)
```

```python
await rate.entrypoint(requests, model='gpt-4-1106-preview', max_tokens=1, temperature=0)
```

```python
for i, row in data.iterrows():
    if row['Answer'].endswith('。'):
        data.loc[i, 'answer_complete'] = 1
```

```python
for i, row in data.iterrows():
    if row['Answer'] == row['Question']:
        data.loc[i, 'answer_complete'] = 0
```

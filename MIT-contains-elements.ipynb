```python
import pandas as pd
import numpy as np
import rate
```

```python
rate.setkey(0)
```

```python
raw = pd.read_excel('MitAut_20231219.xlsx', 'MIT')
```

```python
raw
```

```python
stacked = raw.set_index(raw.columns[:9].to_list()).stack().reset_index().rename(columns={0:'Answer', 'Item 1': 'Item_1', 'Item 2': 'Item_2'})
stacked
```

```python
first_answer = stacked['level_9']=='Answer'
last_answer = pd.concat([first_answer[1:], pd.Series([True])]).reset_index(drop=True)
stacked.loc[~last_answer, 'Time_left'] = -1
data = stacked.drop(columns='level_9')
data
```

```python
data['Answer'] = data['Answer'].apply(lambda x: x.strip() if x.strip() else np.nan)
data.dropna(subset=['Answer'], inplace=True)
```

```python
q_key = 'In the scenario "{Answer}", the two items "{Item_1}" and "{Item_2}" both play a key role: Yes/No'
q_nokey = 'In the scenario "{Answer}", the two items "{Item_1}" and "{Item_2}" both play a role: Yes/No'
```

```python
def process_key(rq, completion, **kwargs):
    a = completion['choices'][0]['message']['content']
    if a != 'Yes' and 'max_tokens' in kwargs:
        return {'model':'gpt-4-1106-preview', 'temperature':0}
    if a == 'Yes':
        data.loc[rq['i'], 'items_play_key_role'] = a
    else:
        data.loc[rq['i'], 'items_play_key_role'] = 'No'
        data.loc[rq['i'], 'no_key_role_full_answer'] = a
```

```python
def process_nokey(rq, completion, **kwargs):
    a = completion['choices'][0]['message']['content']
    if a != 'Yes' and 'max_tokens' in kwargs:
        return {'model':'gpt-4-1106-preview', 'temperature':0}
    if a == 'Yes':
        data.loc[rq['i'], 'items_play_role'] = a
    else:
        data.loc[rq['i'], 'items_play_role'] = 'No'
        data.loc[rq['i'], 'no_role_full_answer'] = a
```

```python
requests_key = [{'messages': [{'role': 'user', 'content': q_key.format(**row)}], 'process': process_key, 'i': i} for i, row in data.iterrows()]
await rate.entrypoint(requests_key, model='gpt-4-1106-preview', max_tokens=1, temperature=0)
```

```python
requests_nokey = [{'messages': [{'role': 'user', 'content': q_nokey.format(**row)}], 'process': process_nokey, 'i': i} for i, row in data.iterrows()]
await rate.entrypoint(requests_nokey, model='gpt-4-1106-preview', max_tokens=1, temperature=0)
```

```python
data.to_excel('MitAut_20231219_GPT-element-use.xlsx')
```

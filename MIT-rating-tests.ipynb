```python
novelty_checks = pd.read_excel(f'Sanity_check_solutions_V6_Final.xlsx', 'Novelty').rename(columns={
    'Q_no':'MIT_Question_no', 'Item1': 'Item_1', 'Item2': 'Item_2'})
novcheck = pd.concat([
    novelty_checks.drop('Loser: NOT novel solutions(JPN)', axis=1).rename(columns={'Winner: Novel solution(JPN)': 'Answer'}).assign(winner=True),
    novelty_checks.drop('Winner: Novel solution(JPN)', axis=1).rename(columns={'Loser: NOT novel solutions(JPN)': 'Answer'}).assign(winner=False)
]).sort_values('MIT_Question_no').reset_index(drop=True)
await method.rate(novcheck, measures=['novelty'])
```

```python
diff_nov = []
for i, winner in novcheck[novcheck.winner].iterrows():
    loser = novcheck[~novcheck.winner & (novcheck.MIT_Question_no==winner.MIT_Question_no)].iloc[0]
    diff = winner.novelty - loser.novelty
    diff_nov.append(diff)
plt.plot(diff_nov, 'ro')
np.mean(diff_nov)
```

```python
feasibility_checks = pd.read_excel(f'Sanity_check_solutions_V6_Final.xlsx', 'Feasibility').rename(columns={
    'Q_no':'MIT_Question_no', 'Item1': 'Item_1', 'Item2': 'Item_2'})
feascheck = pd.concat([
    feasibility_checks.drop('Loser: NOT feasible solutions(JPN)', axis=1).rename(columns={'Winner: Feasible solution(JPN)': 'Answer'}).assign(winner=True),
    feasibility_checks.drop('Winner: Feasible solution(JPN)', axis=1).rename(columns={'Loser: NOT feasible solutions(JPN)': 'Answer'}).assign(winner=False)
]).sort_values('MIT_Question_no').reset_index(drop=True)
await method.rate(feascheck, measures=['feasibility'])
```

```python
diff_feas = []
for i, winner in feascheck[feascheck.winner].iterrows():
    loser = feascheck[~feascheck.winner & (feascheck.MIT_Question_no==winner.MIT_Question_no)].iloc[0]
    diff = winner.feasibility - loser.feasibility
    diff_feas.append(diff)
plt.plot(diff_feas, 'ro')
np.mean(diff_feas)
```

```python
goal_checks = pd.read_excel(f'Sanity_check_solutions_V6_Final.xlsx', 'Goal').rename(columns={
    'Q_no':'MIT_Question_no', 'Item1': 'Item_1', 'Item2': 'Item_2'})
gcheck = pd.concat([
    goal_checks.drop('Loser: Low Goal Attainemnt solutions(JPN)', axis=1).rename(columns={'Winner: High Goal Attainment solution(JPN)': 'Answer'}).assign(winner=True),
    goal_checks.drop('Winner: High Goal Attainment solution(JPN)', axis=1).rename(columns={'Loser: Low Goal Attainemnt solutions(JPN)': 'Answer'}).assign(winner=False)
]).sort_values('MIT_Question_no').reset_index(drop=True)
await method.rate(gcheck, measures=['goal'])
```

```python
diff_goal = []
for i, winner in gcheck[gcheck.winner].iterrows():
    loser = gcheck[~gcheck.winner & (gcheck.MIT_Question_no==winner.MIT_Question_no)].iloc[0]
    diff = winner.goal - loser.goal
    diff_goal.append(diff)
plt.plot(diff_goal, 'ro')
np.mean(diff_goal)
```

```python
# Filter by the same criteria as for ELO ratings (see MIT-to-json.ipynb)
criteria = {
    'elements_used': answer_data['both_items_included']==3,
    'answer_complete': (answer_data['FeedBack_question_3']=='Y') | (answer_data['Time_left']<0),
    'problem_understood': (answer_data['FeedBack_question_1']=='Y') & (answer_data['FeedBack_question_2']=='Y'),
}
criteria['all'] = reduce(lambda x,y: x&y, criteria.values())
source = answer_data[criteria['all']]
```
